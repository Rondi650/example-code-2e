```python
dial_codes = [                                                  # <1>
    (880, 'Bangladesh'),
    (55,  'Brazil'),
    (86,  'China'),
    (91,  'India'),
    (62,  'Indonesia'),
    (81,  'Japan'),
    (234, 'Nigeria'),
    (92,  'Pakistan'),
    (7,   'Russia'),
    (1,   'United States'),
]
```

```python
country_dial = {country: code for code, country in dial_codes}
country_dial
```

```python
{code: country.upper() 
    for country, code in sorted(country_dial.items())
    if code < 70}
```

```python
from random import shuffle
shuffle(dial_codes)
country_dial = {country: code for code, country in dial_codes}
country_dial
```
